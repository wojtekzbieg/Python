Data cleaning

```python
try:
    df = pd.read_csv("/Users/wojtek/Documents/pokemons.csv")
except:
    df = pd.read_csv("C:/Users/Wojtek/Downloads/pokemons.csv")
```

```python
# 1.Dropping a column
df = df.drop(columns=["Legendary", "No"])
df
```

```python
# 2.Handle missing data
df = df.dropna(subset=["Type2"])
print(df.to_string())
```

```python
df = df.fillna({"Type2": "None"})
df
```

```python
# 3.Fix inconsistent values
df["Type1"] = df["Type1"].replace({"Grass": "GRASS",
                                   "Fire": "FIRE",
                                   "Water": "WATER"})
df
```

```python
# 4.Standarize text
df["Name"] = df["Name"].str.lower()
df
```

```python
# 5.Fix data types
df["Legendary"] = df["Legendary"].astype(bool)
df
```

```python
df["Legendary"] = df["Legendary"].replace({1: "True", 0: "False"})
df
```

```python
# 6.Remove duplicate values
df = df.drop_duplicates()
df
```
