```python
import pandas as pd
```

Series

```python
data = [100, 102, 104, 200, 202]

series = pd.Series(data, index=["a", "b", "c", "d", "e"])
print(series)
```

```python
series.loc["c"] = 204
print(series)
```

```python
series.iloc[0] = 0
print(series)
```

```python
print(series[series >= 200])
```

```python
calories = {"Day 1": 1700, "Day 2": 2100, "Day 3": 1850}

series = pd.Series(calories)

print(series)
print(series.loc["Day 1"])
```

```python
series["Day 3"] += 500
print(series)
```

```python
print(series[series > 2000])
print(series[series <= 2000])
```

DataFrame

```python
data = {
    "Name": ["Spongebob", "Patrick", "Squidward"],
    "Age": [30, 35, 50]
}

df = pd.DataFrame(data, index=["Employee 1", "Employee 2", "Employee 3"])
print(df)
```

```python
print(df.loc["Employee 1"])
print()
print(df.iloc[1])
```

```python
#Add a new column
df["Job"] = ["Cook", "N/A", "Cashier"]
print(df)
```

```python
#Add a new row
new_df = pd.DataFrame([{"Name": "Sandy", "Age": 28, "Job": "Scientist"},
                       {"Name": "Eugene", "Age": 60, "Job": "Manager"}],
                        index=["Employee 4", "Employee 5"])
df = pd.concat([df, new_df])

print(df)
```
